```python


# Function to plot a bar chart for population in a specific year
def plot_population_bar_chart(df, year):
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Country Name', y=str(year), data=df.sort_values(by=str(year), ascending=False), palette='viridis')
    plt.title(f'Population Distribution in {year}')
    plt.xlabel('Country Name')
    plt.ylabel('Population')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

```python
# Function to plot a histogram for population in a specific year
def plot_population_histogram(df, year):
    plt.figure(figsize=(12, 8))
    sns.histplot(df[str(year)], bins=20, kde=True, color='blue')
    plt.title(f'Population Distribution in {year}')
    plt.xlabel('Population')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
# Example usage
plot_population_bar_chart(df, 2022)  # Bar chart for year 2022
plot_population_histogram(df, 2022)  # Histogram for year 2022
```

## Identifying top 10 and bottom 10 countries

```python
def top_bottom_countries(df, year, n=10):
    df_year = df[['Country Name', str(year)]].dropna().sort_values(by=str(year))
    top_countries = df_year.tail(n)
    bottom_countries = df_year.head(n)
    return top_countries, bottom_countries
```

```python
def plot_population_bar_chart(df, year, n=10):
    top_countries, bottom_countries = top_bottom_countries(df, year, n)
    
    plt.figure(figsize=(14, 8))
    combined = pd.concat([top_countries, bottom_countries])
    sns.barplot(x='Country Name', y=str(year), data=combined, palette='viridis')
    plt.title(f'Top {n} and Bottom {n} Countries by Population in {year}')
    plt.xlabel('Country Name')
    plt.ylabel('Population')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

```python
def plot_population_histogram(df, year, n=10):
    top_countries, bottom_countries = top_bottom_countries(df, year, n)
    combined = pd.concat([top_countries, bottom_countries])
    
    plt.figure(figsize=(12, 8))
    sns.histplot(combined[str(year)], bins=10, kde=True, color='blue')
    plt.title(f'Population Distribution for Top {n} and Bottom {n} Countries in {year}')
    plt.xlabel('Population')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
# Example usage
plot_population_bar_chart(df, 2022)  # Bar chart for year 2022
plot_population_histogram(df, 2022)  # Histogram for year 2022
```

```python

# Filter for male and female population data
df_male = df[df['Series Name'] == 'Population, male']
df_female = df[df['Series Name'] == 'Population, female']

# Pivot data to have years as columns and countries as rows
df_male = df_male.pivot(index='Country Name', columns='Series Name', values='2022')
df_female = df_female.pivot(index='Country Name', columns='Series Name', values='2022')

# Rename columns for clarity
df_male = df_male.rename(columns={'Population, male': '2022 Male'})
df_female = df_female.rename(columns={'Population, female': '2022 Female'})

# Merge the male and female data into one DataFrame
df_combined = pd.merge(df_male, df_female, left_index=True, right_index=True, how='inner')

print("Transformed Data (head):")
print(df_combined.head())
```

```python
# Function to find top 10 countries with highest male population
def top_countries_by_male_population(df, year, n=10):
    df_sorted = df.sort_values(by=f'{year} Male', ascending=False)
    top_countries = df_sorted.head(n)
    return top_countries

```

```python
# Function to find bottom 10 countries with lowest male population
def bottom_countries_by_male_population(df, year, n=10):
    df_sorted = df.sort_values(by=f'{year} Male', ascending=True)
    bottom_countries = df_sorted.head(n)
    return bottom_countries
```

```python
# Function to find top 10 countries with highest female population
def top_countries_by_female_population(df, year, n=10):
    df_sorted = df.sort_values(by=f'{year} Female', ascending=False)
    top_countries = df_sorted.head(n)
    return top_countries
```

```python
# Function to find bottom 10 countries with lowest female population
def bottom_countries_by_female_population(df, year, n=10):
    df_sorted = df.sort_values(by=f'{year} Female', ascending=True)
    bottom_countries = df_sorted.head(n)
    return bottom_countries
```

```python
# Function to plot a bar chart for top or bottom countries by population
def plot_population_bar_chart(df, column, title, n=10):
    plt.figure(figsize=(14, 8))
    sns.barplot(x=df.index, y=column, data=df, palette='viridis')
    plt.title(title)
    plt.xlabel('Country Name')
    plt.ylabel('Population')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

# Example usage for visualization
year = 2022
n = 10
```

```python
# Top and bottom 10 countries by male population
top_male = top_countries_by_male_population(df_combined, year, n)
bottom_male = bottom_countries_by_male_population(df_combined, year, n)

plot_population_bar_chart(top_male, f'{year} Male', f'Top {n} Countries by Male Population in {year}')
plot_population_bar_chart(bottom_male, f'{year} Male', f'Bottom {n} Countries by Male Population in {year}')
```

```python
# Top and bottom 10 countries by female population
top_female = top_countries_by_female_population(df_combined, year, n)
bottom_female =bottom_countries_by_female_population(df_combined, year, n)

plot_population_bar_chart(top_female, f'{year} Female', f'Top {n} Countries by Female Population in {year}')
plot_population_bar_chart(bottom_female, f'{year} Female', f'Bottom {n} Countries by Female Population in {year}')
```
